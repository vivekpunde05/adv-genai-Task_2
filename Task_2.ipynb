```python
# Problem 1: Unique Words in a Sentence 
# Taking input from user
sentence = input("Enter sentence: ")

# Splitting sentence into words
words = sentence.split()

# Converting list to set to remove duplicates
unique_words = set(words)

# Printing results
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

```python
# Problem 2: Highest Salary from Employee Data
# Store employee names and salaries in a dictionary
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Initialize variables
highest_employee = ""
highest_salary = 0

# Use for loop to compare salaries
for name, salary in employees.items():
    if salary > highest_salary:
        highest_salary = salary
        highest_employee = name
# Display the result
print("Highest Salary:", highest_employee, "-", employees[highest_employee])
```

```python
# Problem 3: Find Maximum and Minimum Values
# List of numbers
numbers = [45, 22, 89, 10, 66]

# Initialize max and min with first element
maximum = numbers[0]
minimum = numbers[0]

# Loop through the list to compare values
for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

# Display results
print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

```python
#Problem 4: Count Products Above a Price Threshold
# List of product prices
prices = [450, 1200, 899, 1500, 300]

# Initialize counter
count = 0

# Check each price
for price in prices:
    if price > 1000:
        count += 1

# Display result
print("Products above 1000:", count)
```

```python
#Problem 5: Calculate Attendance Percentage
# Attendance list (P = Present, A = Absent)
attendance = ["P", "P", "A", "P", "P"]

# Count number of presents
present_count = attendance.count("P")

#total number of days 
total_days = len(attendance)

#calculate persentage 
percentage = (present_count / total_days) *100

# Display result
print("Attendance Percentage:", percentage)
```

```python
#Problem 6: Remove Duplicate Phone Numbers
# List of phone numbers (with duplicates)
phone_numbers = [9876543210, 9123456789, 9876543210]

# Convert list to set to remove duplicates
unique_numbers = set(phone_numbers)

# Display unique phone numbers
print("Unique phone numbers:", unique_numbers)
```

```python
#Problem 7: Count Character Frequency
# Input string
text = "pythonp"

# Create empty dictionary
frequency = {}

# Count each character
for char in text:
    if char in frequency:
        frequency[char] += 1
    else:
        frequency[char] = 1

# Display result
print("Character Frequency:", frequency)
```

```python
#Problem 8: Convert List to Tuple
# List of numbers
numbers = [10, 20, 30]

# Convert list to tuple
result = tuple(numbers)

# Display tuple
print("Tuple:", result)
```

```python
#Problem 9: Check if a Key Exists in a Dictionary
# Employee dictionary
employees = {"Ravi": 75000, "Anita": 68000}

# Check if key exists
if "Ravi" in employees:
    print("Employee exists")
else:
    print("Employee does not exist")
```

```python
#Problem 10: Calculate Average Marks
# List of marks
marks = [70, 80, 65, 75]

# Calculate total marks
total = 0
for mark in marks:
    total = total + mark

# Calculate average
average = total / len(marks)

# Display result
print("Average Marks:", average)
```
